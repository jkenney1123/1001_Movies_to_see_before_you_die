```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker
import plotly.express as px
```

```python
raw1001 = pd.read_csv("1001 Movies You Must See Before You Die (All Editions Combined).csv")
```

```python
raw1001
```

```python
fig, ax = plt.subplots(nrows=2,ncols=3, figsize = (15,8))
#ax[0,1].hist(raw1001.loc[:,'IMDb Rating'],bins = 20)
#ax[0,1].set_xlabel("IMDb Rating")
#ax[0,1].set_ylabel("Counts")
#ax[0,1].set_title("Histogram of IMDb Ratings of the 1001 Movies")

# Set theme and color palette similar to ggplot2
sns.set_theme(style="whitegrid")

sns.histplot(raw1001.loc[:,'IMDb Rating'],kde = True,bins = 20, ax= ax[0,0])
sns.histplot(raw1001.loc[:,'Runtime (mins)'],kde = True,bins = 20, ax= ax[0,1])
sns.histplot(raw1001.loc[:,'Num Votes'],kde = True,bins = 20, ax= ax[0,2])
sns.barplot(raw1001.loc[:,'Year'].value_counts(), ax= ax[1,0])
sns.boxplot(raw1001[raw1001['Title Type'].isin(["Short", "Movie"])],x = "IMDb Rating", y = "Title Type", ax= ax[1,1],hue = "Title Type",legend=False,showmeans = True, meanline = True,flierprops={'marker': 'D', 'label': "Title",'markersize': 4, 'markerfacecolor': 'black','alpha' : 0.5})
#px.box(raw1001[raw1001['Title Type'].isin(["Short", "Movie"])].loc[:,['Title Type','IMDb Rating']],x = "IMDb Rating", y = "Title Type", ax= ax[1,1])

ax[0,0].set_title("Histogram of IMDb Ratings of the 1001 Movies")
ax[0,1].set_title("Histogram of Runtime (mins) of the 1001 Movies")
ax[0,2].set_title("Histogram of Num Votes of the 1001 Movies")
ax[1,0].set_title("# of Movies for each Year in the 1001 Movies")
ax[1,0].xaxis.set_major_locator(ticker.MaxNLocator(nbins=10))
plt.tight_layout()
plt.show()
```
